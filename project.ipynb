# 2017


In the economies of Kyrgyzstan's main trading partners, there was a gradual recovery of economic activity. In the first 3 months of 2017, the gradual recovery of the Russian economy continued, with inflation slowing at an accelerated pace, approaching the target level of 4 %. In Kazakhstan, economic growth was supported by the implementation of structural reforms and state programs, as well as growth in the production and export of raw materials. Inflationary pressures in Kazakhstan have been slowing since the beginning of the year and are approaching target benchmarks. In the economy of the Republic of Belarus, there was a recovery of growth against the backdrop of improving external demand, and inflation has slowed down due to improvements in the domestic currency market situation. In the economy of Armenia, there is a recovery of economic activity, with actual inflation slightly accelerating but still remaining below the target benchmark.

High rates of economic growth were observed. Economic growth was mainly driven by a substantial increase in industrial production and a significant influx of remittances into the country, which had a positive impact on domestic consumption. All major sectors of the economy contributed positively to the economic growth, except for the construction sector.

In particular, production at the "Kumtor" enterprise in 2017 amounted to 127.4 thousand ounces, increasing by 47.5 percent compared to 2016 (86.4 thousand ounces). There was also growth in the production of refined petroleum products (by 45.9 percent), food products (by 37.0 percent), and clothing (by 31.9 percent). The services sector grew due to expansion in the trade and transportation sectors, while there was a contraction in the information and communication sector.

The significant easing of monetary conditions undertaken in 2016 contributed to neutralizing deflationary pressures. Since the beginning of 2017, inflation dynamics have taken an upward trajectory and remained within expected values. Considering this, monetary policy conditions were maintained during the first quarter of 2017. The behavior of market participants was largely explained by the preservation of a sufficiently high level of excess liquidity in the banking system. Money market rates fluctuated within the interest rate corridor set by the National Bank. Taking into account the forecasted inflation dynamics for the medium term, as well as in order to support the real sector of the economy, the National Bank continued to conduct bank refinancing operations during the first quarter of 2017.

In the first quarter of 2017, inflation began to recover, primarily due to the rise in prices of fruit and vegetable products, driven by the low base of the previous year. However, inflation remained at a relatively low level to some extent due to the high supply of certain goods.

# 2018

The trend of economic activity recovery in the main trading partners of the Kyrgyz Republic persisted. In Russia, some economic slowdown is expected, along with temporary inflation acceleration due to the increase in VAT. In Kazakhstan, economic growth was driven by production activity in all sectors of the economy, while the downward trend in the tenge exchange rate raised inflation expectations in the country. Economic growth in Belarus was fueled by increased production in all sectors of the economy except agriculture, as well as favorable external conditions and stimulating state policies; inflation accelerated against the backdrop of expanding domestic demand and weakening national currency. Positive economic growth rates persisted in Armenia, with inflation remaining below target. Assumptions regarding a gradual slowdown in China's economic growth perspective remained unchanged.
Positive economic growth rates persisted. Economic activity in the economies of Kyrgyzstan's main trading partners and a positive influx of remittances into the republic had a positive impact on demand.


```python
fig, ax=plt.subplots(2, 2)
ax[0,0] .plot([3.2, 4.5, 5.4], [2017.25, 2017.5, 2018])
ax[0,1] .plot([3.9, 4.6, 11.0], [2017.25, 2017.5, 2018])
ax[1,0] .plot([2.1, 5.3, 9.8], [2017.25, 2017.5, 2018])
ax[1,1] .plot([2.1, 3.7, 9.2], [2017.25, 2017.5, 2018])
plt.show()
```

Based on the results of 2018, the average monthly nominal wage of one worker amounted to 16,218 soms, increasing by 5.4 percent compared to 2017. The most significant increase in the growth rate of the average monthly wage was observed in construction (11.0 percent), in the information and communication sector (9.8 percent), and in the sphere of public administration and defense, compulsory social security (9.2 percent).

The maintenance of low inflation in the country, coupled with relatively stable prices in global food markets, created conditions for maintaining a stimulative direction of monetary policy in the fourth quarter of 2018. Tactical decisions in monetary policy were made considering the high level of excess liquidity in the banking system, which tended to increase towards the end of the reporting period due to seasonal factors. This situation led the National Bank to conduct sterilization operations during the fourth quarter of 2018.

There was increased activity in the interbank credit market, both due to an increase in market participants and transaction volumes. Overall, money market interest rates remained within the interest rate corridor set by the National Bank, continuing to shift towards the key rate.

In order to improve the transmission mechanism of monetary policy and develop the interbank money market, the National Bank twice raised the lower bound of the interest rate corridor (the overnight deposit rate) from 0.25 to 1.5 percent during the fourth quarter of 2018. This led to a reduction in the spread between money market rates and the key rate.

In 2018, inflation in Kyrgyzstan was characterized by low growth rates of prices and did not exceed the target benchmarks set by the National Bank. The inflationary backdrop remained stable, which was further supported by favorable developments in the global food market.


# 2019

Economic activity in the main trading partners of the Kyrgyz Republic remained moderate. During the reporting period, economic growth in Russia accelerated against the backdrop of increased investment spending in national projects. Inflation in Russia continued to slow down and ended the year below the target set by the Central Bank of Russia. Economic growth in Kazakhstan was mainly driven by the expansion of domestic demand, which influenced price growth. Economic growth in Belarus remained at the level of the previous quarter, with primary support coming from domestic demand; inflation slowed down by the end of the year, falling below the target set by the National Bank of Belarus. In Armenia, economic activity reached its highest level since 2011; inflation remained at the level of the previous quarter, slowing down overall for the year 2019. In China, economic growth reached its lowest levels in recent years and was supported by government stimulus measures. The pace of economic growth in the United States was determined at the level of the third quarter, decreasing for the entire year to the level of 2016.

The increase in food prices on global markets, mainly due to rising prices for vegetable oils and meat, posed certain inflationary risks for Kyrgyzstan. During the reporting period, there was an increase in oil prices on the market against the backdrop of positive news about the agreement on the first phase of the trade deal between the United States and China. Gold prices overall had an upward trend.

In January-December 2019, positive economic growth rates were maintained. The main driver of the economy during the reporting period was the industrial sector, including the positive impact on the country's economic development from the increased production of basic metals at enterprises involved in the development of the Kumtor deposit.

In conditions of moderate inflation, the stimulative direction of monetary policy aimed at supporting economic activity in the country was continued during the fourth quarter of 2019. Tactical measures of monetary policy were formulated, on one hand, considering the need to provide the banking system with long-term resources through monetary instruments for refinancing priority sectors of the economy, including regional development, and on the other hand, for managing the existing excess liquidity in the system, a significant portion of which is short-term and unevenly distributed.

Previous efforts to improve the functioning of the monetary policy mechanism influenced the dynamics of short-term interest rates in the money market during the reporting period, bringing them closer to the National Bank's policy rate, demonstrating their consistency and gradual reduction of the gap with the key rate.

In Kyrgyzstan, inflation moderately increased from the second half of 2019. The level of inflation remained within the expectations of the National Bank, not exceeding the target benchmarks for the medium term, and reflecting the dynamics of the global food market.

# 2020


The economic activity in the main trading partners of the Kyrgyz Republic was moderately recovering. The GDP decline in Russia in 2020 turned out to be more moderate than expected by both the government and international analysts, while inflation accelerated due to short-term pro-inflationary factors and exceeded the target set by the Bank of Russia. In Kazakhstan, there remained a negative trend in economic development in the fourth quarter of 2020, although with an improvement compared to the previous two quarters. In the period under review, economic activity in Belarus remained at the level of the previous quarter. At the same time, inflation accelerated, surpassing the forecast of the National Bank of the Republic of Belarus, mainly due to the weakening of the Belarusian ruble against the backdrop of increased inflationary and devaluation expectations in the economy. In Armenia, economic activity contracted due to the negative consequences of the COVID-19 pandemic and the military conflict in Karabakh, while inflation accelerated. In the fourth quarter of 2020, China, showing high rates of recovery after the COVID-19 outbreak, became the only global power to avoid an economic downturn by the end of 2020. In 2020, the US economy contracted for the first time since the 2008 financial crisis and demonstrated the largest decline since World War II.

The price indices on the global food markets reached their highest levels in the last three years, which posed certain inflationary risks for Kyrgyzstan. During the period under consideration, there was a recovery in oil prices on the market. The gold market demonstrated stability with a tendency to rise towards the end of the year.
In 2020, for the first time in 25 years, the economic growth rates showed a significant decline due to restrictive measures to prevent the spread of the coronavirus infection and because of the unstable political situation in the country.

Based on the preliminary results of the third quarter of 2020, the decline in domestic demand on an annual basis amounted to -8.9 percent. The decrease was mainly associated with a reduction in gross accumulation (-21.1 percent), primarily due to a decrease in changes in inventories of material circulating assets (-42.7 percent). In January-November 2020, the average nominal monthly wage of one worker amounted to 18.0 thousand soms, increasing by 8.3 percent compared to the same period in 2019. The most significant growth in wage growth rates was observed in the education sector (20.6 percent) and administrative and support activities (20.3 percent), water supply, sanitation, waste management, and remediation activities (14.6 percent). Meanwhile, the growth in real wages in January-November 2020 compared to the same period in 2019 was 2.2 percent.

The measures introduced in March 2020 due to the imposition of emergency and martial law regimes in some regions of the republic had a significant impact on reducing economic activity, while the closure of borders with neighboring countries complicated the delivery of goods. Additionally, there has been a noted negative impact on economic growth from rallies, strikes, and production stoppages during the post-election period.
Throughout 2020, the monetary base significantly increased due to changes in the structure of government operations resulting from a noticeable gap between the revenue and expenditure parts of the state budget.
At the end of 2020, inflation sharply accelerated and exceeded the expectations of the National Bank, although the annual average inflation rate did not exceed the target benchmarks for the medium term.

# 2021

The growth of business activity globally and in the trading partners of the Kyrgyz Republic is slowing down against the backdrop of an active phase of economic recovery and the introduction of new restrictive measures against COVID-19 outbreaks. Persistent constraints in global logistics chains, along with the consequences of stimulus measures aimed at minimizing the pandemic's impact, are driving high inflationary pressures in all countries worldwide.

Based on the results of the first 10 months of 2021, the growth of real GDP was 1.6 percent. Excluding enterprises involved in the development of the Kumtor mine, the GDP growth rate was 4.1 percent. The positive dynamics of economic recovery are supported by the increasing volume of remittances into the country. From January to September 2021, the net inflow of remittances from individuals increased by 19.0 percent, amounting to approximately 1.6 billion USD.
Amid rising inflationary pressures and growing inflation expectations, the National Bank continued to tighten monetary conditions. Short-term interest rates in the money market increased following the key rate and were formed within the interest rate corridor. Tactical decisions in the field of monetary policy were made under conditions of significant growth in excess liquidity in the banking sector, with a predominance of shorter-term funds. In these circumstances, the short segment of the money market was characterized by the highest activity among participants. 
Overall, the monetary policy implemented by the National Bank increased the yield and attractiveness of assets denominated in the national currency.
It is expected that the further rise in global food prices will be one of the main drivers of inflation in the Kyrgyz Republic. Additionally, the increase in administered and regulated prices and tariffs may contribute further to inflation.

# 2022


Significant changes in the external environment in 2022 impacted the development of the Kyrgyz Republic's economy and, consequently, its monetary policy. Throughout the reporting year, the heightened inflationary background in Kyrgyzstan reflected the influence of both external and internal factors.

In this context, the National Bank's monetary policy measures were aimed at curbing the monetary component of inflation and minimizing the negative impact of external shocks on the domestic economy. A comprehensive set of measures utilizing a wide range of monetary policy tools was adopted. These included raising the National Bank's key interest rate from 8.00 percent to 14.00 percent at the beginning of 2022, followed by a reduction to 13.00 percent at the end of 2022 as price dynamics relatively stabilized.

The timely measures helped contain the acceleration of inflation. The average annual inflation rate from 2018 to 2022 was 7.0 percent, which was within the National Bank's medium-term target range.

To regulate liquidity levels in the banking system, the National Bank increased the volume of sterilization operations to withdraw excess liquidity. Additionally, to cover short-term liquidity gaps and ensure the stable functioning of the banking system, the National Bank provided commercial banks with credit resources.

In 2022, there was increased demand for both cash and non-cash foreign currency in the domestic foreign exchange market. This demand was primarily driven by the situations in the main trading partner countries of the Kyrgyz Republic and disruptions in supply chains of cash US dollars into the country. Consequently, the National Bank conducted foreign exchange interventions at certain times to smooth out sharp exchange rate fluctuations and stabilize the foreign exchange market.
The impact of the pandemic and political instability
In 2020, Kyrgyzstan's economy experienced a significant downturn, for the first time in 25 years, due to restrictive measures aimed at preventing the spread of the coronavirus infection, as well as political instability in the country.


# 2023

The global economy maintains slow growth amid tight monetary policy conditions and restrained lending. While global inflation is decelerating, core inflation, which excludes volatile energy and food prices, remains elevated.

In the United States, consumer spending supports the economy, whereas in China, weak domestic and export demand negatively impact economic activity. All member states of the Eurasian Economic Union (EAEU) have recorded positive GDP growth dynamics.
Prices on the global food market have continued to decrease over the past twelve months, while in many countries, domestic prices remain higher than the previous year. The main factors contributing to this trend include deteriorating macroeconomic conditions in countries, conflicts, high prices for agricultural resources, and volatility of national currencies against the US dollar or euro.

There is a positive trend in oil prices on the global market due to a decrease in supply. Amid ongoing geopolitical risks and high inflation in the global gold market, there has been mixed price dynamics during the reporting period.
As of the results of the first half of 2023, the economic growth rate in the Kyrgyz Republic has slowed down due to low production figures at the Kumtor deposit. The main contribution to economic growth comes from the service sector. There is a continuing trend of reducing the inflow of remittances from individuals into the country.

During the second quarter of 2023, the external economic conditions that directly and indirectly impact the economy of the Kyrgyz Republic remained uncertain and ambiguous. The situation on the external financial markets continued to be volatile, leading to some volatility in the domestic currency market as well.

Overall, the inflationary backdrop in the Kyrgyz Republic gradually weakened. This was contributed to by the stabilization of the situation in the global food markets and previously implemented monetary policy measures. However, internal factors such as the increase in certain administered prices (tariffs) and the gradual expansion of domestic demand remained inflationary.

In response to these inflationary risks, the National Bank maintained appropriate monetary policy conditions. To limit the monetary component of inflation, the National Bank actively increased its sterilization operations given the high level of excess liquidity in the banking system. Short-term interest rates in all segments of the money market remained close to the key rate. As a result of the measures taken by the National Bank, the situation in the money and currency markets remained relatively stable.
In the second quarter of 2023, there was a slowdown in the year-on-year growth rate of the Consumer Price Index (CPI) in Kyrgyzstan due to a decrease in the rate of increase in prices for food products. However, external economic conditions continued to be uncertain, with geopolitical risks persisting, which still posed upward risks to inflation.

During the analysis of inflation dynamics in Kyrgyzstan for the specified period, it was established that the country faced various challenges affecting the level of the price level. Despite periods of relative stability, inflationary processes in the country remained quite volatile overall. This was due to both internal and external factors, such as changes in global prices for energy resources, food, as well as political instability in the region.

In analyzing the inflation regulation policy, it became apparent that the Central Bank of Kyrgyzstan actively applied various monetary instruments to contain inflationary pressures. However, the effectiveness of these measures remained variable due to insufficient coordination with fiscal policy and limited capabilities of the Central Bank.

Additionally, during the study, it was found that inflationary processes have a significant impact on the economic well-being of the population. Price increases affect the purchasing power of the population, especially those with low income levels.
